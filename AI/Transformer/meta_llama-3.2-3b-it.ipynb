```python
# 下载模型的 tokenizer 和 model 两个物件
from transformers import AutoTokenizer, AutoModelForCausalLM

# 只要更换 model_id 就可以换成其他模型（HuggingFace 上的模型）
# 例如 meta-llama/Llama-3.2-3B-Instruct, google/gemma-3-4b-it 等
model_id = "unsloth/Llama-3.2-3B-Instruct"

# tokenizer 记录了模型所使用的 token，可以理解为模型的字典
# model 存储了模型的参数
# 下载过程可能需要数分钟，请耐心等待，如果你选择下载更大的模型，所需时间也会相对更长
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id)
```

```python
# 查看这个模型有多少个 token
# https://huggingface.co/docs/transformers/main_classes/tokenizer
print("语言模型有多少个不同的 Token 可以选择：", tokenizer.vocab_size)
```

```python
# 打印 token
# token_id 范围从 0 到 tokenizer.vocab_size-1 (可穷举所有 token 的编号）
for token_id in range(0, 93):
    token = tokenizer.decode(token_id)
    print("Token 编号 %d (长度 %d) 是：%s" % (token_id, len(token), token))
```

```python
# 文本与 token 编码解码
# 加上 add_special_tokens=False 可以避免加上代表起始的符号
text = "大家好，我们一起来学习"
tokens = tokenizer.encode(text, add_special_tokens=False)
print("encode", text, "->", tokens)
text_after_decode = tokenizer.decode(tokens)
print("Decode", tokens, "->", text_after_decode)
```
